```python
import sympy as sp

def taylor(f, x, a, absolute_error, point, ej):
    """
    Compute the Taylor series expansion of a function f around a point a until the absolute error at a given point is below a given value.

    Args:
        f: The function to approximate.
        x: The variable of the function.
        a: The point around which to expand the Taylor series.
        absolute_error: The maximum absolute error allowed for the approximation at the given point.
        point: The point at which to check the absolute error of the approximation.
        max_terms: Maximum number of terms in the Taylor series.

    Returns:
        The Taylor series expansion of f until the absolute error at the given point is below the given value.
    """
    n = 0
    i = 0
    data = []
    puntos = np.linspace(-2, 2, 100)
    taylor = f.series(x, a, n=n).removeO()
    approximation = taylor.subs(x, point)  # Evaluate the initial approximation at the point a
    resor = [f.subs(x, punto) for punto in puntos]
    plt.plot(puntos, resor, label='Funcion original')
    while True:
        approximation_at_point = approximation.subs(x, point)
        error = sp.N(sp.Abs(sp.N(f.subs(x, point)) - approximation_at_point))  # Compute the absolute error at the given point
        data.append([n, sp.N(approximation), error, sp.N(f.subs(x, point))])

        if error <= absolute_error:
            break
        n += 1
        taylor = f.series(x, a, n=n).removeO()
        approximation = taylor.subs(x, point)  # Update the approximation
        resultados = [taylor.subs(x, punto) for punto in puntos]
        plt.plot(puntos, resultados, label=f'Polinomio de orden{i}')
        i+=1

# Convertir la lista de resultados en un array de numpy si es necesario
        resultados_array = np.array(resultados)
    df = pd.DataFrame(data, columns=['Orden Poly','Aprox','Error','Valor_Verdadero'])
    plt.legend()
    plt.title('Ejercicio'+ej)
    plt.show()
    return df,taylor

### Tema 1
# a)
x = sp.symbols('x')
f = sp.sin(x)  # Define la función
a = sp.pi  # Punto alrededor del cual expandir la serie
absolute_error = 0.0001/100  # Error absoluto máximo permitido
point = 2  # Punto en el que se verificará el error absoluto de la aproximación
taylor_expansion = taylor(f, x, a, absolute_error, point, ' a')
print(taylor_expansion)
# b)
f = sp.cos(2*x)
a = sp.pi/2
absolute_error = 0.00016/100
point = -1
taylor_expansion = taylor(f, x, a, absolute_error, point, ' b')
print(taylor_expansion)
# c)
point = 0
f = sp.exp(x)
a = 1
absolute_error = 0.000052/100
taylor_expansion = taylor(f, x, a, absolute_error, point, ' c')
print(taylor_expansion)
# d)
point = -0.361
f = 11 - x
a = 0
absolute_error = 0.000021/100
taylor_expansion = taylor(f, x, a, absolute_error, point, ' d')
print(taylor_expansion)
# e)
f = sp.exp(-5*x)
point=0.2
a = 0
absolute_error = 0.005/100
taylor_expansion = taylor(f, x, a, absolute_error, point, ' e')
print(taylor_expansion)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def diferencia_progresiva(f, x, h):
    return (f.subs(x, x + h) - f) / h

def diferencia_regresiva(f, x, h):
    return (f - f.subs(x, x - h)) / h

def diferencia_centrada(f, x, h):
    return (f.subs(x, x + h) - f.subs(x, x - h)) / (2 * h)
# Función para calcular las aproximaciones de las derivadas y los errores relativos
def errorDF(f, x, punto, hs):
    diccionario = {'h': [], 'progresiva': [], 'regresiva': [], 'centrada': [], 'real': [],
                   'e_progresiva': [], 'e_regresiva': [], 'e_centrada': []}

    for h in hs:
        # Calcular aproximaciones de las derivadas
        df_progresiva = diferencia_progresiva(f, x, h).subs(x, punto)
        df_regresiva = diferencia_regresiva(f, x, h).subs(x, punto)
        df_centrada = diferencia_centrada(f, x, h).subs(x, punto)
        df_exacta = sp.diff(f, x).subs(x, punto)

        # Calcular errores relativos
        e_progresiva = sp.Abs((df_exacta - df_progresiva) / df_exacta)
        e_regresiva = sp.Abs((df_exacta - df_regresiva) / df_exacta)
        e_centrada = sp.Abs((df_exacta - df_centrada) / df_exacta)

        # Agregar a diccionario
        diccionario['h'].append(h)
        diccionario['progresiva'].append(df_progresiva)
        diccionario['regresiva'].append(df_regresiva)
        diccionario['centrada'].append(df_centrada)
        diccionario['real'].append(df_exacta)
        diccionario['e_progresiva'].append(e_progresiva)
        diccionario['e_regresiva'].append(e_regresiva)
        diccionario['e_centrada'].append(e_centrada)

    # Crear DataFrame
    df = pd.DataFrame(diccionario)
    return df
def graficar_errores(df):
    plt.figure(figsize=(10, 6))
    plt.scatter(df['h'], df['e_progresiva'], label='Error Progresiva', color='red')
    plt.scatter(df['h'], df['e_regresiva'], label='Error Regresiva', color='blue')
    plt.xscale('log')
    plt.yscale('log')
    plt.title('Errores Relativos en Función de h')
    plt.xlabel('h')
    plt.ylabel('Error Relativo')
    plt.legend()
    plt.grid(True)
    plt.show()
def calcular_aproximaciones_derivadas(f, x, punto, hs):
    df_exacta = sp.diff(f, x)  # Derivada exacta de la función
    d2f_exacta = sp.diff(df_exacta, x)  # Segunda derivada exacta de la función

    # Funciones para evaluar
    f_func = sp.lambdify(x, f, 'numpy')
    df_exacta_func = sp.lambdify(x, df_exacta, 'numpy')
    d2f_exacta_func = sp.lambdify(x, d2f_exacta, 'numpy')

    # Valores para el gráfico
    x_vals = np.linspace(punto - 4, punto + 4, 400)
    f_vals = f_func(x_vals)
    df_exacta_vals = df_exacta_func(x_vals)
    d2f_exacta_vals = d2f_exacta_func(x_vals)

    # Gráfico de la función original y su aproximación de la primera derivada
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x_vals, f_vals, label='Función Original')
    plt.title('Función Original y Aproximación de la Primera Derivada')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)

    for h in hs:
        # Aproximaciones de la primera derivada
        df_progresiva = diferencia_progresiva(f, x, h)
        df_regresiva = diferencia_regresiva(f, x, h)
        df_centrada = diferencia_centrada(f, x, h)

        # Valores para el gráfico de las aproximaciones de la primera derivada
        df_progresiva_vals = sp.lambdify(x, df_progresiva, 'numpy')(x_vals)
        df_regresiva_vals = sp.lambdify(x, df_regresiva, 'numpy')(x_vals)
        df_centrada_vals = sp.lambdify(x, df_centrada, 'numpy')(x_vals)

        # Gráfico de las aproximaciones de la primera derivada
        plt.plot(x_vals, df_progresiva_vals, label=f'Progresiva (h={h})')
        #plt.plot(x_vals, df_regresiva_vals, label=f'Regresiva (h={h})')
        #plt.plot(x_vals, df_centrada_vals, label=f'Centrada (h={h})')

    plt.legend()

    # Gráfico de la función original y su aproximación de la segunda derivada
    plt.subplot(1, 2, 2)
    plt.plot(x_vals, f_vals, label='Función Original')
    plt.title('Función Original y Aproximación de la Segunda Derivada')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)

    if d2f_exacta.is_constant():
        plt.plot(x_vals,np.array([d2f_exacta_vals]*len(x_vals)), color='r', linestyle='--', label='Aproximación de la Segunda Derivada')

    else:
        for h in hs:
            # Aproximaciones de la segunda derivada
            d2f_progresiva = diferencia_progresiva(df_exacta, x, h)
            d2f_regresiva = diferencia_regresiva(df_exacta, x, h)
            d2f_centrada = diferencia_centrada(df_exacta, x, h)

            # Valores para el gráfico de las aproximaciones de la segunda derivada
            d2f_progresiva_vals = sp.lambdify(x, d2f_progresiva, 'numpy')(x_vals)
            d2f_regresiva_vals = sp.lambdify(x, d2f_regresiva, 'numpy')(x_vals)
            d2f_centrada_vals = sp.lambdify(x, d2f_centrada, 'numpy')(x_vals)

            # Gráfico de las aproximaciones de la segunda derivada
            plt.plot(x_vals, d2f_progresiva_vals, label=f'Progresiva (h={h})')
            #plt.plot(x_vals, d2f_regresiva_vals, label=f'Regresiva (h={h})')
            #plt.plot(x_vals, d2f_centrada_vals, label=f'Centrada (h={h})')

    plt.legend()
    plt.tight_layout()
    plt.show()
```
